```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('712 data.csv')
df
```

```python
df_Mens = df[df['FITTING_ROOM_AREA']=='Mens']
```

```python
df_Mens_pv = df_Mens.pivot_table(values = 'AVERAGE_FR_OCCUPANCY_PCT', index = 'DAY_OF_THE_WEEK', columns = 'TIME_PERIOD')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (10,6))
plt.title('Fitting Room Occupancy (%) in Manchester Trafford During Week 24')
sns.heatmap(df_Mens_pv, cmap = 'Oranges', linecolor='white',linewidths=1, annot = True,cbar_kws={'format': '%.0f%%'})
plt.xlabel('Time Period')
plt.ylabel('Day of the Week')
```
